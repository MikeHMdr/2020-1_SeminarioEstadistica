```python
# Importamos las bibliotecas necesarias
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api   as sm
```

Leemos los datos desde un archivo separado por comas

```python
#datos = pd.read_csv("datos_ajuste_1.csv")                  # Datos comunes y corrientes
#datos = pd.read_csv("datos_ajuste_1_dato_atipico.csv")           # Datos con 1 dato atípico (pero no influyente)
datos = pd.read_csv("datos_ajuste_1_dato_influyente.csv")       # DAatos con 1 dato atípico e influyente
datos.tail()
```

Graficamos el conjunto de datos con una gráfica de dispersión (scatter-plot)

```python
X, Y = datos["X"], datos["Y"] 

plt.figure( figsize=(16,6) )                           # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                              # Añadimos el titulo
plt.scatter( X, Y , label = "Datos de prueba" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                    # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                    # Modificamos la etiqueta del eje Y
plt.grid()                                             # Añadimos una cuadricula/rendija 
plt.legend()                                           # Añadimos un cuadro con la leyenda
plt.show()                                             # Indicamos que queremos mostrar la gráfica
```

Realizamos la regresión desde "cero".

Construimos a mano nuestra matriz de diseño, los residuales, la descomposición de la suma de cuadrados, la matriz sombrero, entre otros.

```python
# Creamos nuestra regresión desde cero
M          = np.matrix( sm.add_constant(X) )                  # Matriz de Diseño
n          = M.shape[0]                                       # Número de observaciones
p          = M.shape[1]                                       # Número de covariables (incluyendo beta0)
betas      = ( M.T * M ) ** (-1) * M.T * np.matrix( Y ).T     # Vector Columna de Betas Estimadas
Y_est      = np.array( M * betas ).reshape( n, )              # Arreglo (unidimensional) de las Y Estimadas
residuales = np.array(Y) - np.array(Y_est)                    # Arreglo (unidimensional) de los residuales 

# Descomposición de la suma de cuadrados :: SCT = SCE + SCR
Y_barra = np.mean(np.array(Y))
SCT = np.sum(  ( np.array(Y) - Y_barra )**2  )
SCE = np.sum(  ( np.array(Y) - np.array(Y_est) )**2  )
SCR = np.sum(  ( np.array(Y_est) - Y_barra )**2  )

# El error que ocurria en la descomposición de cuadrados se daba cuando definiamos Y_est := M * betas
# Esto porque era un vector Columna de las Y Estimadas y cuando haciamos SCE = np.sum(  ( np.array(Y) - np.array(Y_est) )**2  )
# No realizaba la diferencia elemento a elemento. También aplica para SCR. 
# Si desean ver lo que hacia revisen la última celda del Notebook

# R2 = Coeficiente de determinación
R2 = SCR / SCT 
R2 = 1 - SCE / SCT

# R2 ajustada 
R2_adj = 1 - ( SCE/(n-p) ) / ( SCT/(n-1) )

# Matriz sombrero e influencia de cada observación
# NOTA :: Es importante que M sea objeto tipo Matrix. En caso contrario, marcará error.
H                         = M * ( M.T * M ) ** (-1) * M.T    # Matriz Sombrero (H de Hat)
influencia                = np.array( H.diagonal() )         # Influencia de cada observacion (leverage)
observaciones_influyentes = np.array(list(range(n)))[ np.array( influencia > 2*np.mean(influencia) ).reshape( n,) ]
# Una regla empirica muy usada es considerar observaciones unfluyentes si superan 2 veces la influencia promedio

# Realizamos un grafico de dispersión ( Scatter-Plot )
plt.figure( figsize=(16,6) )                                       # Indicamos el tamaño de la gráfica
plt.title("Scatter Plot")                                          # Añadimos el titulo
plt.scatter( X, Y , label = "Datos Observados" )                   # Añadimos la 'capa' de puntos 
plt.plot( X, Y_est , color="red", label = "Recta de Regresión" )   # Añadimos la 'capa' de puntos 
plt.xlabel("Eje X")                                                # Modificamos la etiqueta del eje X
plt.ylabel("Eje Y")                                                # Modificamos la etiqueta del eje Y
plt.grid()                                                         # Añadimos una cuadricula/rendija 
plt.legend()                                                       # Añadimos un cuadro con la leyenda
plt.show()                                                         # Indicamos que queremos mostrar la gráfica

# Mostramos algunos valores importantes
print("Estadísticas Importantes ------------------------------------")
print("SCT                 :: ", SCT)
print("SCE + SCR           :: ", SCE + SCR)
print("SCE                 :: ", SCE)
print("SCR                 :: ", SCR)
print("R2                  :: ", R2) 
print("R2 adj              :: ", R2_adj)
print("Influencia promedio :: ", np.mean(influencia) )
print("Indices de las observaciónes que superan 2 veces la influencia promedio (En Python los indices empiezan en cero!!!)")
print( observaciones_influyentes )
print("-------------------------------------------------------------")
```

**EXTRA**

Ver que sucedió en la descomposición de la suma de cuadrados por Y_est

```python
# Creamos nuestra regresión desde cero
M          = np.matrix( sm.add_constant(X) )                  # Matriz de Diseño
n          = M.shape[0]                                       # Número de observaciones
p          = M.shape[1]                                       # Número de covariables (incluyendo beta0)
betas      = ( M.T * M ) ** (-1) * M.T * np.matrix( Y ).T     # Vector Columna de Betas Estimadas
# Vease que se modificó la definición de Y_est a un vector columna
Y_est      = M * betas                                        # Vector Columna de las Y Estimadas

print("Dimensiones de Y_est")         
print(Y_est.shape)                    # Esto es un arreglo en dos dimensiones
print("Dimensiones de np.array(Y_est)")         
print(np.array(Y_est).shape)          # Esto sigue siendo un arreglo en dos dimensiones
print("Dimensiones de np.array(Y)") 
print(np.array(Y).shape)              # Esto es un vector unidimensional

# Si tratamos de sacar los residuales veremos que hay una diferencia en el tipo de dato / estructura / dimensiones
# Es decir, si hacemos la diferencia de np.array(Y) - np.array(Y_est) la resta de no se hace elemento a elemento
# pues no tienen la misma dimensión. La única resta correcta es en el primer elemento
caso_erroneo  = np.array(Y) - ( M * betas )
caso_correcto = np.array(Y) - np.array( M * betas ).reshape( n, )
print("")
print("Caso erroneo")
print("Dimensiones :: ", caso_erroneo.shape )
print("Valores     :: ", caso_erroneo )
print("")
print("Caso correcto")
print("Dimensiones :: ", caso_correcto.shape )
print("Valores     :: ", caso_correcto )
```
