# Ejemplo de Árbol de Decisión

Importamos las bibliotecas necesarias

```python
import pandas            as pd    # Biblioteca para trabajar con DataFrames
import numpy             as np    # Biblioteca para trabajar con àlgebra lineal
import matplotlib.pyplot as plt   # Biblioteca para realizar gràficas
from sklearn.metrics         import accuracy_score, confusion_matrix   # Funciones para medir la precisión
from sklearn.model_selection import cross_val_score                    # Funciones para selección de modelos
from sklearn.tree            import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree # Funciones de clase Tree
# https://scikit-learn.org/stable/modules/tree.html
import warnings
warnings.filterwarnings("ignore")
```

Leemos el archivo `iris.csv`

```python
print("Vista previa del dataset:")
datos_iris = pd.read_csv("iris.csv")
datos_iris.head()
```

Información general del dataset.
Tipo de dato de cada variable, las dimensiones de la tabla, así como el tamaño en memoria

```python
print('Información del dataset:')
datos_iris.info()
```

Resumen de cada variable

```python
print('Descripción del dataset:')
datos_iris.describe()
```

Separamos nuestras variables explicativas y nuestra variable respuesta.

```python
# Datos para realizar el ajuste
matriz_diseño = datos_iris.values[:,0:4]
y_real        = datos_iris.values[:,4]

print("Matriz de Diseño:")
print(matriz_diseño[0:5,:])
print("")
print("Y Real")
print(y_real[0:5])
```

El árbol de decisión es una clase con varios parámetros. A continuación se muestran los parámetros por defecto.

```python
# Modelo de Árbol de Decisión con parámetros por defecto
DecisionTreeClassifier()
```

Ajustamos nuestro modelo de árbol de decisión con los parámetros de defecto.

Unicamente colocamos una semilla `(random_state)` para replicar los resultados

```python
# Modelo de Arbol de Decisión con parámetros por defecto, unicamente que están especificados
modelo_arbol = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=None, max_features=None,
                max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, min_samples_leaf=1,
                min_samples_split=2, min_weight_fraction_leaf=0.0, presort=False, random_state=0, splitter='best')
modelo_arbol.fit(matriz_diseño, y_real)
modelo_arbol
```

Graficamos nuestra árbol de decisión

```python
# Graficamos el arbol
plt.figure( figsize=(16,8) )                                      # Indicamos el tamaño de la gráfica
plot_tree(modelo_arbol, filled=True, rounded=True, precision=2 )  # Graficamos el árbol de decisión 
plt.show()                                                        # Indicamos que queremos mostrar la gráfica
```

Dentro del modelo de árbol de decisión, tenemos diferentes atributos/propiedades.

Entre los cuales están las clases a estimar, número de clases, el criterio de división, la importancia de las variables entre otras cosas.


```python
# Algunos detalles del arbol de decisión
print( modelo_arbol.classes_ )
print( modelo_arbol.n_classes_ ) 
print( modelo_arbol.criterion ) 
print( modelo_arbol.feature_importances_ ) 
```

Estimamos nuestra variable respuesta y posteriormente medimos la precisión del modelo

```python
# Estimamos nuestro valor Y
y_estimada = modelo_arbol.predict( matriz_diseño )

print("Y Estimada")
print(y_estimada[0:5])

# Realizamos nuestra matriz de confusión y calculamos la precisión del modelo
print("")
print("Matriz de Confusión")
print( confusion_matrix(y_real, y_estimada) )

print("")
print("Precisión del modelo")
print( accuracy_score(y_real, y_estimada) * 100  )
```

```python
# Grafica de la importancia de las variables
plt.figure( figsize=(16,8) )
plt.barh( datos_iris.columns[0:4] , modelo_arbol.feature_importances_ )
plt.show()
```
